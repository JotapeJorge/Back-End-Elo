```python
#Unites the different datasets

import pandas as pd

Brasileiro = pd.read_csv('data/dataBRA.csv')
Brasileiro23 = pd.read_csv('data/dataBRA23.csv')
SérieB = pd.read_csv('data/dataBRB.csv')
SérieB23 = pd.read_csv('data/dataBRB23.csv')
SérieC = pd.read_csv('data/dataBRC.csv')
SérieC23 = pd.read_csv('data/dataBRC23.csv')
Copa = pd.read_csv('data/dataCBR.csv')
Copa23 = pd.read_csv('data/dataCBR2023.csv')

# Merge dataframes and sort by date
frames = [Brasileiro, Brasileiro23, SérieB, SérieB23, SérieC, SérieC23, Copa, Copa23]
df = pd.concat(frames)
df = df.drop_duplicates()
df['Data'] = pd.to_datetime(df['Data'], infer_datetime_format=True, errors='coerce')
df = df.sort_values(by='Data')
df.to_csv('data/dataELO.csv', index=False)
df
```

```python
#I use this to make sure that was no duplicates in the DF
import Levenshtein

def are_similar(s1, s2, threshold=0.85):
    distance = Levenshtein.distance(s1.lower(), s2.lower())
    similarity = 1 - distance / max(len(s1), len(s2))
    return similarity >= threshold

# Your DataFrame containing the 'Casa' column
# Assuming you have already converted the "Data" column to pandas datetime format
df['Data'] = pd.to_datetime(df['Data'])

# Generate a list of unique values from the 'Casa' column and sort them
unique_values = df['Fora'].unique()
unique_values_sorted = sorted(unique_values)

# Find semiduplicates using Levenshtein distance with a threshold of 0.9 (90% similarity)
semiduplicates = []

for i in range(len(unique_values_sorted)):
    for j in range(i + 1, len(unique_values_sorted)):
        if are_similar(unique_values_sorted[i], unique_values_sorted[j]):
            semiduplicates.append((unique_values_sorted[i], unique_values_sorted[j]))

print("Semiduplicates:")
for pair in semiduplicates:
    print(f"{pair[0]} and {pair[1]} are similar.")
```
